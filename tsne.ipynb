```python
import torch
import sklearn
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
x = torch.load('E.tensor').cpu().numpy()
```

```python
year = list(range(69, 80))
movie = list(range(1014,1028))
author = list(range(1077,1085))+list(range(17160,17169))
country = [38496,41882,42284,40513,40928,43011,39849,42558,41303,38958,40420,39686]
genre = [7125,27896,42225,42226,42422,42423,38993,42362,40285,40286]
animal = [39001,39013,4058,39075,41479,11185,39846,38723,38724,42816]
idx = year+movie+author+country+genre+animal
color = [0]*len(year) + [1]*len(movie) + [2]*len(author) + [3] * len(country) + [4] * len(genre) + [5]*len(animal)
name = ['year']*len(year) + ['movie']*len(movie) + ['author']*len(author) + ['country'] * len(country) + ['genre'] * len(genre) + ['animal']*len(animal)
X = x[idx]
```

```python
X_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=2).fit_transform(X)
```

```python
scatter = plt.scatter(X_embedded[:,0],X_embedded[:,1],c=color,)
plt.legend(handles=scatter.legend_elements()[0], labels=['year','movie','author','country','genre','animal'])
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.set_xticks([0,1,2,3,4,5,6],[1, 2, 5, 10, 20, 50, 100])
ax1.plot([0, 1, 2, 3, 4], [0.09, 0.12, 0.25, 0.31, 0.28], 'g-')
ax2.plot([0, 1, 2, 3, 4, 5, 6],[1, 3/2, 6/2, 11/2, 21/2, 51/2, 101/2], 'b-')
ax1.set_xlabel('Corruption Number')
ax1.set_ylabel('Training Performance', color='g')
ax2.set_ylabel('Training Time', color='b')
```
